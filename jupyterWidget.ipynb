```python
from ipywebrtc import CameraStream, ImageRecorder #realtimeCommunication-->chatting
```

```python
#help(CameraStream)
```

```python
camera=CameraStream.facing_user(audio=False)
#help(ImageRecorder)
```

```python
#hook up the image came with stream
image_recorder=ImageRecorder(stream=camera)

image_recorder.recording=True
image_recorder.download()
type(image_recorder.image)
```

```python
from ipywidgets import Image
import PIL.Image
import io
```

```python
img=PIL.Image.open(io.BytesIO(image_recorder.image.value))
display(img)
```

```python
import cv2 as cv
```

```python
face_cascade=cv.CascadeClassifier("face.xml")
eye_cascade=cv.CascadeClassifier("eye.xml")
```

```python
cv_img=cv.imread("img/faces.png")
faces=face_cascade.detectMultiScale(cv_img)
faces
```

```python
#faces.tolist()[0]
```

```python
from PIL import Image,ImageDraw
pil_img=Image.open("img/faces.png")
drawing=ImageDraw.Draw(pil_img)
```

```python
for x,y,w,h in faces:
    drawing.rectangle((x,y,x+w,y+h),outline="red")
display(pil_img) 
```

```python
def show_rect(faces):
    pil_img=Image.open("img/faces.png")
    drawing=ImageDraw.Draw(pil_img)
    for x,y,w,h in faces:
        drawing.rectangle((x,y,x+w,y+h),outline="red")
    display(pil_img)
```

```python
cv_img_bin=cv.threshold(cv_img,120,255,cv.THRESH_BINARY)[1]
```

```python
faces=face_cascade.detectMultiScale(cv_img_bin)
show_rect(faces)
```

```python
faces=face_cascade.detectMultiScale(cv_img,1.05)
show_rect(faces)#only showed eyes
```
