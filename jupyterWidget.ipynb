```python
from ipywebrtc import CameraStream, ImageRecorder #realtimeCommunication-->chatting
```

```python
#help(CameraStream)
```

```python
camera=CameraStream.facing_user(audio=False)
#help(ImageRecorder)
```

```python
#hook up the image came with stream
image_recorder=ImageRecorder(stream=camera)

image_recorder.recording=True
image_recorder.download()
type(image_recorder.image)
```

```python
from ipywidgets import Image
import PIL.Image
import io
```
