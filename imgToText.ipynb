```python
img=img.convert('L')
img.save('img/greyscale_0045.jpg')
display(img)
```

```python
text=pytesseract.image_to_string(Image.open("img/greyscale_0045.jpg"))
print(text)
```

```python
blackNwhite = img.point(lambda x: 0 if x < 200 else 255, '1')
blackNwhite.save("img/blackNwhite_0045.jpg")
display(blackNwhite)
```
